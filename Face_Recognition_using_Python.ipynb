```python
#convert to garyscale
gray_image=(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY))
gray_image
```

```python
gray_image.shape
```

```python
smaller_image_size=cv2.resize(img,(100,100))
cv2_imshow(smaller_image_size)
```

```python
ret,thresh_binary=cv2.threshold(gray_image,175,255,cv2.THRESH_BINARY)

ret,thresh_binary_inv=cv2.threshold(gray_image,175,255,cv2.THRESH_BINARY_INV)

ret,thresh_trunc=cv2.threshold(gray_image,175,255,cv2.THRESH_TRUNC)

ret,thresh_tozero=cv2.threshold(gray_image,175,255,cv2.THRESH_TOZERO)

ret,thresh_tozero_inv=cv2.threshold(gray_image,175,255,cv2.THRESH_TOZERO_INV)
```

```python
import matplotlib.pyplot as plt
#displaying the different thresholding styles usig OpenCV
names=['Original Name','BINARY','THRESH_BINARY','THRESH_TRUNC','THRESH_TOZERO','THRESH_TOZERO_INV']

images=gray_image,thresh_binary,thresh_binary_inv,thresh_trunc,thresh_tozero,thresh_tozero_inv

for i in range(6):
  plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')

  plt.title(names[i])

  plt.xticks([]),plt.yticks([])
```

```python
face_cascade=cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
```

```python
faces=face_cascade.detectMultiScale(gray_image,1.1,4)

#draw rectangle around the faces
for(x,y,w,h) in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)

#diplay the output
cv2_imshow(img)
cv2.waitKey()
```
